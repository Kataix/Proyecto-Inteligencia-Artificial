```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import networkx as nx
```

```python
estaciones = pd.read_csv('DataSetMetroStationMexico.csv', sep="\t")
G = nx.Graph()
distancia = []
```

```python
for a in estaciones['Longitud de interestación']:
    a = a.replace(",","")
    a = int(a)
    distancia.append(a)
```

```python
estaciones = estaciones.drop(columns=['Longitud de interestación'])
estaciones['Distancia'] = distancia
print(estaciones)
```

```python
for origen, destino, tramo in zip(estaciones['Origen'], estaciones['Destino'], estaciones['Distancia']):
    G.add_edge(origen, destino, weight = tramo)
fig, ax = plt.subplots(figsize = (40,30))
print(nx.dijkstra_path(G,"Cerro de la Estrella","La Paz"))

nx.draw(G, with_labels=True, node_color='red', font_size=10, node_size = 100, )
plt.show()
```
